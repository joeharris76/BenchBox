# Trend Analysis

Plot performance evolution for a single platform across multiple runs.

```python
from datetime import datetime

from benchbox.core.visualization import NormalizedQuery, NormalizedResult, ResultPlotter

runs = [
    NormalizedResult(
        benchmark="TPC-H",
        platform="duckdb",
        scale_factor=1,
        execution_id="duckdb-2024",
        timestamp=datetime(2024, 6, 1),
        total_time_ms=10_800,
        avg_time_ms=360,
        success_rate=1.0,
        cost_total=9.0,
        queries=[NormalizedQuery(query_id="Q1", execution_time_ms=360, status="SUCCESS")],
    ),
    NormalizedResult(
        benchmark="TPC-H",
        platform="duckdb",
        scale_factor=1,
        execution_id="duckdb-2025",
        timestamp=datetime(2025, 6, 1),
        total_time_ms=9_700,
        avg_time_ms=323,
        success_rate=1.0,
        cost_total=9.5,
        queries=[NormalizedQuery(query_id="Q1", execution_time_ms=323, status="SUCCESS")],
    ),
    NormalizedResult(
        benchmark="TPC-H",
        platform="duckdb",
        scale_factor=1,
        execution_id="duckdb-2026",
        timestamp=datetime(2026, 6, 1),
        total_time_ms=8_900,
        avg_time_ms=297,
        success_rate=1.0,
        cost_total=10.0,
        queries=[NormalizedQuery(query_id="Q1", execution_time_ms=297, status="SUCCESS")],
    ),
]

plotter = ResultPlotter(runs, theme="light")
plotter.generate_all_charts(
    output_dir="charts/trends",
    template_name="trends",
    formats=("png", "html"),
)
```
